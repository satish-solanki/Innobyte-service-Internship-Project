```python
# import libries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the csv file  from Drive
df = pd.read_csv('/content/drive/MyDrive/Data Science/INNOBYTE INTERNSHIP/Amazon Sale Report.csv')
```

```python
# first look of data
df
```

```python
# checking shape of the data
df.shape
```

```python
# data info
df.info()
```

```python
# check the duplicate value in data
df.duplicated().sum()
```

```python
# remove the duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
# unique value in each column
unique_values = df[df.columns].nunique()
unique_values = unique_values[unique_values < 2]
unique_values
```

```python
# column which have only one unique value that are not usefull for analysis, they are not giving any anylysis
# remove the columns with one unique value
df.drop(['ship-country','fulfilled-by','currency','Fulfilment','Sales Channel','ship-service-level'], inplace=True, axis=1)
df.shape
```

```python
# check for null values
df.isnull().sum()
```

```python
# null values are very less as comparision to full data set
# drop the null values
df.dropna(inplace=True)
df.info()
```

```python
#changed the data type of requered column
df['Date'] = pd.to_datetime(df['Date'])

df['ship-postal-code'] = df['ship-postal-code'].astype('int')
df['B2B'] = df['B2B'].astype('bool')

```

```python
#make a copy of datset
new_df = df.copy()
new_df.head(10)
```

```python
#check the discription about data
numerice_col = ['Amount','Qty']
new_df[numerice_col].describe()
```

```python
# Amount and Qty column which have 0 values
Amount_zero = new_df[new_df['Amount'] == 0]
Qty_zero = new_df[new_df['Qty'] == 0]
Qty_zero
```

```python
# all the amount value to 0 if the stutus is cancelled

new_df.loc[new_df['Status'] == 'Cancelled', 'Amount'] = 0
new_df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('Amazone_sales_Cleaned_data.csv', index=False)

print("Data saved to output.csv")
```
